```python
import datetime
moviedata['Open']=moviedata.Open.iloc[:].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d'))
```

```python
import pickle
with open('googletrendss.txt', 'rb') as fp1:
            google=pickle.load(fp1)        
with open('googletrendss2.txt', 'rb') as fp2:
            google2=pickle.load(fp2)
with open('googletrends.txt', 'rb') as fp3:
            google22=pickle.load(fp3)
with open('googletrends 22.txt', 'rb') as fp4:
            google222=pickle.load(fp4)
```

```python
google3=google+google2+google22+google222
len(google3)
```

```python
google_dic={}
for i in google3:
    if isinstance(i[1], int):
        google_dic[i[0]]=i[1]
    else:
        google_dic[i[0]]=i[1][0]
```

```python
len(google_dic)
```

```python
result=[]
for i in moviedata["Title"]:
    if i in google_dic.keys():
        result.append(google_dic[i])
    else:
        result.append(None)
```

```python
len(google_dic)
```

```python
len(moviedata["Title"])
```

```python
len(result)
```

```python
moviedata['googletrends']=pd.Series(result)
```

```python
moviedata[moviedata['googletrends'].isnull()].to_csv('/Users/diwu/Desktop/Group Project/1.csv')
```
